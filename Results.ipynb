## Simulations
- 100 packets
- packet size = 2 bytes - 20 bytes

```python
results = {}
algorithms = ["broadcast", "CCBR"]
topologies = [1,2,3,4]
packet_sizes = [2,10,20]
nodes = [8,12,8,12]
packets = [100,100,100,100]

for algorithm in algorithms:
    print(algorithm)
    results[algorithm] = list()

    for topology in topologies:
        print("\nTOPOLOGY:", topology)

        for packet_size in packet_sizes:
            filename = algorithm + "/simulations/logs/" + str(packet_size) +"/loglistener_sim_" + str(topology) + ".txt"
            file = open(filename, "r")
            received = list()
            sent_by_nodes = list()
            sent_by_sink = list()
            broadcasted_by_nodes = list()
            received_by = []

            for i in range(0, nodes[topology-1]+1):
                received_by.append(list())

            for line in file:
                if ('SinkC: Broadcasting packet 101' in line):
                    break
                if ('SinkC: Packet sent' in line):
                    sent_by_sink.append(line)
                if ('NodeC: Packet sent' in line):
                    sent_by_nodes.append(line)
                if ('NodeC: Broadcasting packet' in line):
                    broadcasted_by_nodes.append(line)
                if ('NodeC: Received the new packet' in line):
                    received.append(line)
                for i in range(2, nodes[topology-1]+2):
                    if ('ID:' + str(i) + '	NodeC: Received the new packet' in line):
                        received_by[i-2].append(line)

            print("Packet size: %d" % packet_size)
            print("\tPackets sent by sink: %d" % len(sent_by_sink))
            print("\tPackets sent by nodes: %d" % len(sent_by_nodes))
            print("\tPackets broadcasted by nodes: %d" % len(broadcasted_by_nodes))
            print("\tReceived packets: %d\n" % len(received))

            total = 0
            print("\tNode \tNum \t%")
            for i in range(0, nodes[topology-1]):
                print("\t%d  \t%d \t%.2f" % (i+2, len(received_by[i]), (len(received_by[i])/packets[topology-1])*100))
                total = total + len(received_by[i])

            print("\t-----------------------------")
            print("\n\tReceived packets total: %d/%d, %.2f%%" % (total, (packets[topology-1]*nodes[topology-1]), (total/(packets[topology-1]*nodes[topology-1]))*100))
            print("\tPackets sent by nodes: %d" % len(sent_by_nodes))
            
            results[algorithm].append([topology, packet_size, (total/(packets[topology-1]*nodes[topology-1])*100), len(sent_by_nodes)])

            print("\n")
```

```python
delta = list()

from tabulate import tabulate
for algorithm in algorithms:
    print("\t\t\t", algorithm)
    print(tabulate(results[algorithm], headers=["Topology", "Packet size", "% received", "Packets"]))
    print("\n")

for i in range(0,len(results["broadcast"])):
    delta.append([results["broadcast"][i][0], results["broadcast"][i][1], (results["broadcast"][i][2] - results["CCBR"][i][2]), (results["broadcast"][i][3] - results["CCBR"][i][3])])

print("\n\t\tBroadcast VS CCBR")
print(tabulate(delta, headers=["Topology", "Packet size", "Δ Received", "Δ Packets"]))
```
